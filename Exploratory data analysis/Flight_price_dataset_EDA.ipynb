```python
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
#reading dataset from local system
df = pd.read_excel('flight_price.xlsx')
df.head()
```

```python
#info about the data(works only for numerical column)
df.info()
```

```python
#statistics of the data
df.describe()
```

```python
#first working with Date_of_journey column
df['Date_of_Journey']
```

```python
#spliting date, month and year

#for date
df['date'] = df['Date_of_Journey'].str.split('/').str[0]

#for month
df['month'] = df['Date_of_Journey'].str.split('/').str[1]

#for year
df['year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df['date'], df['month'], df['year']
```

```python
#converting splited date into numerical values
df['date']= df['date'].astype(int)
df['month']=df['month'].astype(int)
df['year']= df['year'].astype(int)
```

```python
df.info()
```

```python
df['arrival_time'] = df['Arrival_Time'].str.split(' ').str[0]
df['arrival_time']
```

```python
df.head()
```

```python
df['arrival_hour'] = df['arrival_time'].str.split(':').str[0]
df['arrival_minute'] = df['arrival_time'].str.split(':').str[1]
df.head()
```

```python
#working with categorical features
df['Total_Stops']
```

```python
df['Total_Stops'].unique()
```

```python
#counting null value
df[df['Total_Stops'].isnull()]
```

```python
df['Total_Stops'] = df['Total_Stops'].map({
    'non-stop':0,
    '1 stop':1,
    '2 stops':2,
    '3 stops':3,
    '4 stops':4,
    np.nan:1
})
df['Total_Stops'].unique()
```

```python
#convert categorical feature in numarical using one-hot-encoding
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder()
```

```python
encoder.fit_transform(df[['Airline']]).toarray()
```

```python
Airline_encoded = pd.DataFrame(encoder.fit_transform(df[['Airline']]).toarray(), columns=encoder.get_feature_names_out())
```
