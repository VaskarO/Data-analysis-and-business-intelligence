```python
df_bios_new = df_bios.copy()
df_bios_new['first_name'] = df_bios_new['name'].str.split(' ').str[0]
df_bios_new['first_name'].head()
```

```python
df_bios_new["Height_category"] = df_bios['height_cm'].apply(lambda x: 'Short' if x <165 else ('Average' if x<165 else 'Tall')
)
df_bios_new.head()
```

```python
def athlete_category(row):
    if row['height_cm']<175 and row['weight_kg']<70:
        return 'light weight'
    elif row['height_cm']<185 or row['weight_kg']<= 80:
        return 'middle weight'
    else:
        return 'heavy weight'

df_bios['Category']= df_bios.apply(athlete_category, axis=1)
df_bios['Category'].head()
```
